```python
###### Python program to create a simple GUI
# calculator using Tkinter

# import everything from tkinter module 
from tkinter import *
from math import *

# globally declare the expression variable
expression = " "

# Function to update expression
# in the text entry box 
def press(num) :
    # point out the global expression variable 
    global expression 
    
    # concatenation of string 
    expression = expression + str(num)
    
    # update the expression by using set method 
    equation.set(expression)
    
# Function to evaluate the final expression 
def equalpress() :
    # Try and except statement is used
    # for handling the errors like zero 
    # division error etc
    
    # Put that code inside the try block
    # which may generate the error
    try :
        
        global expression
        
        # eval function evaluate the expression 
        # and str function convert the result 
        # into string
        total = str(eval(expression))
        
        equation.set(total)
        
        # initialize the expression variable 
        # by empty string
        expression = " "
    
    # if error is generated than handle 
    # by except block
    except : 
        
        equation.set(" error ")
        expression = " "
        
# Function to clear the contents 
# of text entry box
def clear() :
    global expression 
    expression = " "
    equation.set( " " )
    
def backspace():
    global expression
    expression = equation.set.get()[:-1]
    equation.set.delete(0, END)
    equation.set.insert(0, expression)
    
# Function to square root
# of text entry box
def numsqrt():
    global expression
    sqrt = math.sqrt()
    equation.set( " " )
    expression = " "
    
    
# Driver code
if __name__ == "__main__" :
    # create a GUI window
    gui = Tk()
    
    # set the bg Color of GUI window
    gui.configure(background = "Honeydew3")
    
    # set the title of GUI window
    gui.title("Calculator")
    
    # set the configuration of GUI window
    gui.geometry("433x285")
    
    # StringVar() is the variable class
    # we create an instance of this class
    equation = StringVar()
    
    # create the text entry box for 
    # showing the expression
    expression_field = Entry(gui, textvariable = equation, bg = "Honeydew3", fg = "black")
    #expression_field = Text(gui, width = 30, height = 3, background = "yellow", foreground = "blue")
    
    # grid method is used for placing the 
    # widgets at respective positions
    # in table like structure
    expression_field.grid(columnspan = 4, ipadx = 155, ipady = 10)
    
    equation.set('0')
    
    # Create Buttons and place at a particular
    # location inside the root window.
    # when user press the button, the command
    # or function affiliated to that button is executed
    # Button: 1
    button1 = Button(gui, text = '1', fg = 'white', bg = 'grey', #font = ('bold'), 
                    command = lambda: press(1), height = 2, width = 14)
    button1.grid(row = 2, column = 0)
    
    # Button: 2
    button2 = Button(gui, text = '2', fg = 'white', bg = 'grey',
                    command = lambda: press(2), height = 2, width = 14)
    button2.grid(row = 2, column = 1)
    
    # Button: 3
    button3 = Button(gui, text = '3', fg = 'white', bg = 'grey',
                    command = lambda: press(3), height = 2, width = 14)
    button3.grid(row = 2, column = 2)
    
    # Button: +
    buttonadd = Button(gui, text = '+', fg = 'white', bg = 'grey',
                    command = lambda: press("+"), height = 2, width = 14)
    buttonadd.grid(row = 2, column = 3)
    
    # Button: 4
    button4 = Button(gui, text = '4', fg = 'white', bg = 'grey',
                    command = lambda: press(4), height = 2, width = 14)
    button4.grid(row = 3, column = 0)
    
    # Button: 5
    button5 = Button(gui, text = '5', fg = 'white', bg = 'grey',
                    command = lambda: press(5), height = 2, width = 14)
    button5.grid(row = 3, column = 1)
    
    # Button: 6
    button6 = Button(gui, text = '6', fg = 'white', bg = 'grey',
                    command = lambda: press(6), height = 2, width = 14)
    button6.grid(row = 3, column = 2)
    
    # Button: -
    buttonsub = Button(gui, text = '-', fg = 'white', bg = 'grey',
                    command = lambda: press("-"), height = 2, width = 14)
    buttonsub.grid(row = 3, column = 3)
    
    # Button: 7
    button7 = Button(gui, text = '7', fg = 'white', bg = 'grey',
                    command = lambda: press(7), height = 2, width = 14)
    button7.grid(row = 4, column = 0)
    
    # Button: 8
    button8 = Button(gui, text = '8', fg = 'white', bg = 'grey',
                    command = lambda: press(8), height = 2, width = 14)
    button8.grid(row = 4, column = 1)
    
    # Button: 9
    button9 = Button(gui, text = '9', fg = 'white', bg = 'grey',
                    command = lambda: press(9), height = 2, width = 14)
    button9.grid(row = 4, column = 2)
    
    # Button: *
    buttonmul = Button(gui, text = '*', fg = 'white', bg = 'grey',
                    command = lambda: press("*"), height = 2, width = 14)
    buttonmul.grid(row = 4, column = 3)
    
    # Button: Clear
    buttonc = Button(gui, text = 'C', fg = 'white', bg = 'grey',
                    command = clear, height = 2, width = 14)
    buttonc.grid(row = 6, column = 3)
    
    # Button: 0
    button0 = Button(gui, text = '0', fg = 'white', bg = 'grey',
                    command = lambda: press(0), height = 2, width = 14)
    button0.grid(row = 5, column = 1)
    
    # Button: =
    buttoneq = Button(gui, text = '=', fg = 'white', bg = 'grey',
                    command = equalpress, height = 2, width = 14)
    buttoneq.grid(row = 5, column = 2)
    
    # Button: /
    buttondiv= Button(gui, text = '/', fg = 'white', bg = 'grey',
                    command = lambda: press("/"), height = 2, width = 14)
    buttondiv.grid(row = 5, column = 3)
    
    # Button: .
    buttondec = Button(gui, text = '.', fg = 'white', bg = 'grey',
                    command = lambda: press("."), height = 2, width = 14)
    buttondec.grid(row = 5, column = 0)
    
    # Button: bs
    buttonbs = Button(gui, text = 'Backspace', fg = 'white', bg = 'grey',
                    command = backspace, height = 2, width = 14)
    buttonbs.grid(row = 6, column = 2)
    
    # Button: %
    buttonper = Button(gui, text = '%', fg = 'white', bg = 'grey',
                    command = lambda: press("/100*"), height = 2, width = 14)
    buttonper.grid(row = 6, column = 1)
    
    # Button: root
    buttonroot = Button(gui, text = 'Root', fg = 'white', bg = 'grey',
                    command = lambda: press("sqrt"), height = 2, width = 14)
    buttonroot.grid(row = 6, column = 0)
    
    # Button: bracketsleft
    buttonbracl = Button(gui, text = '(', fg = 'white', bg = 'grey',
                    command = lambda: press("("), height = 2, width = 14)
    buttonbracl.grid(row = 7, column = 0)
    
    # Button: bracketright
    buttonbracr = Button(gui, text = ')', fg = 'white', bg = 'grey',
                    command = lambda: press(")"), height = 2, width = 14)
    buttonbracr.grid(row = 7, column = 1)
    
    # start the GUI 
    gui.mainloop()
```
